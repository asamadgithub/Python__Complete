```python
# 0.1*x**5 - 0.2*x**3 +0.1*x -0.2 
# Centeral Difference Method using hand written formula
```

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt

f=lambda x: 0.1*x**5 - 0.2*x**3 +0.1*x -0.2 
x=np.linspace(0,1,11)
#print("Array generated by np.linspace \n",x)


h=0.10
d1_cd= ( f(x+h) - f(x-h) )/(2*h)
d2_cd= ( f(x+h) - 2*f(x) + f(x-h) )/(h**2)

plt.plot(x,f(x),'-k', x,d1_cd,'--b', x,d2_cd,'-.r' )
plt.xlabel('x')
plt.ylabel('y')
plt.legend([ "y","y\'","y\''"])
plt.grid()
plt.show()
```

```python
# Finitite Central Differences using scipy
```

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

f=lambda x: 0.1*x**5 - 0.2*x**3 +0.1*x -0.2 
x=np.linspace(0,1,20)
#print("Array generated by np.linspace \n",x)
h=0.10  # step size

d1_cd= derivative(f,x,h,1)
d2_cd= derivative(f,x,h,2)
#d3_cd,'--k'
plt.plot(x,f(x),'-k', x,d1_cd,'--b',x, d2_cd,'-.r' )
plt.xlabel('x')
plt.ylabel('y')
plt.legend([ "y","y\'","y\''"])
plt.grid()
plt.show()
```
